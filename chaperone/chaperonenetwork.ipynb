```python
# read in chaperone interaction network
# need to verify results against this network

import pandas as pd

# Replace with the correct path to your file
file_path = '/Users/b/1000-networks/mmc5.xls'

# Load the Excel file using xlrd engine and skip the first 8 rows
ppi = pd.read_excel(file_path, sheet_name=None, engine='xlrd', skiprows=8)

```

```python
# Create a dictionary to store protein interactions
protein_interactions = {}

# Iterate over each sheet in the Excel file
for sheet_name, df in ppi.items():
    # Iterate over each row in the sheet
    for index, row in df.iterrows():
        symbol_a = row['Symbol A']
        symbol_b = row['Symbol B']
        
        # Skip if both symbols are identical
        if symbol_a == symbol_b:
            if symbol_a not in protein_interactions:
                protein_interactions[symbol_a] = []
            if symbol_b not in protein_interactions[symbol_a]:
                protein_interactions[symbol_a].append(symbol_b)
            continue
        
        # Add the interaction to the dictionary
        if symbol_a not in protein_interactions:
            protein_interactions[symbol_a] = []
        if symbol_b not in protein_interactions:
            protein_interactions[symbol_b] = []
        
        if symbol_b not in protein_interactions[symbol_a]:
            protein_interactions[symbol_a].append(symbol_b)
        if symbol_a not in protein_interactions[symbol_b]:
            protein_interactions[symbol_b].append(symbol_a)

# Display the dictionary of protein interactions
protein_interactions
```

```python

# Define the evidence list
evidenceList = ["POSITIVE GENETIC", "PHENOTYPIC ENHANCEMENT"]

# Initialize the dictionary for storing interactions
protein_interactions = {}

# Initialize the set for storing all unique proteins
all_proteins = set()

# Function to query BioGRID for interactions
def query_biogrid(protein):
    params = {
        "accesskey": cfg.ACCESS_KEY,
        "format": "json",  # Return results in JSON format
        "geneList": protein,  # Search for the current protein
        "searchNames": "true",  # Search against official names
        "includeInteractors": "true",  # Get any interaction involving either gene
        "includeInteractorInteractions": "true",  # Get interactions between the geneList’s first-order interactors
        "taxId": 9606,  # Limit to Homo sapiens (human)
        "evidenceList": "|".join(evidenceList),  # Exclude these evidence types
        "includeEvidence": "false",  # Evidence to exclude
    }

    r = requests.get(request_url, params=params)
    return r.json() if r.status_code == 200 else {}

# Iterate over each chaperone protein and query BioGRID
for protein in chaperoneList:
        print(f"Querying BioGRID for {protein}...")
        interactions = query_biogrid(protein)
        
        if interactions:  # Check if interactions is not None or empty
            interaction_proteins = []
            
            for interaction_id, interaction in interactions.items():
                symbol_a = interaction['OFFICIAL_SYMBOL_A']
                symbol_b = interaction['OFFICIAL_SYMBOL_B']
                
                # Add the interacting proteins to the list
                if symbol_a != protein:
                    interaction_proteins.append(symbol_a)
                    all_proteins.add(symbol_a)
                if symbol_b != protein:
                    interaction_proteins.append(symbol_b)
                    all_proteins.add(symbol_b)
                
                # Ensure the current protein is added to the set of all proteins
                all_proteins.add(protein)
            
            # Update the dictionary with the interactions for the current protein
            protein_interactions[protein] = interaction_proteins

# Save the dictionary of protein interactions to a file
import json

with open('protein_interactions.json', 'w') as f:
    json.dump(protein_interactions, f, indent=4)

# Save the set of all unique proteins to a file
with open('all_proteins.txt', 'w') as f:
    for protein in all_proteins:
        f.write(protein + '\n')

print("Protein interactions and unique protein set saved.")
```

```python
import json

with open('protein_interactions.json', 'r') as f:
    protein_interactions = json.load(f)

# Initialize a set to keep track of all seen interactions
seen_interactions = set()

# Helper function to create a sorted tuple for an interaction
def interaction_tuple(protein_a, protein_b):
    return tuple(sorted([protein_a, protein_b]))

# Filter and update protein_interactions to remove repeats
for protein, interactions in protein_interactions.items():
    filtered_interactions = []
    for interaction in interactions:
        interaction_pair = interaction_tuple(protein, interaction)
        if interaction_pair not in seen_interactions:
            seen_interactions.add(interaction_pair)
            filtered_interactions.append(interaction)
    protein_interactions[protein] = filtered_interactions

# Save the updated dictionaries back to their respective files
with open('protein_interactions.json', 'w') as f:
    json.dump(protein_interactions, f, indent=4)
```

```python
# Create chaperone_interactions dictionary
chaperone_interactions = {}

# Filter interactions to include only those in chaperoneList
for protein, interactions in protein_interactions.items():
    filtered_interactions = [p for p in interactions if p in chaperoneList]
    if filtered_interactions:  # Only add if there are valid interactions
        chaperone_interactions[protein] = filtered_interactions

# Save the chaperone_interactions dictionary to a file
import json

with open('chaperone_interactions.json', 'w') as f:
    json.dump(chaperone_interactions, f, indent=4)
```

```python
# coexpression layer
import pandas as pd

# Replace with the correct path to your file
file_path = '/Users/b/1000-networks/mmc6.xls'

# Load the Excel file using xlrd engine and skip the first 8 rows
cox_data = pd.read_excel(file_path, sheet_name=None, engine='xlrd', skiprows=8)
```

```python
print(list(cox_data.values())[0])
cox_table = list(cox_data.values())[0]
```

```python

# Filter for correlations greater than 0.7
filtered_cox_table = cox_table[cox_table['Corr (Aging)'] > 0.7]

# Create the dictionary
connections = {}

for _, row in filtered_cox_table.iterrows():
    symbol_a = row['Symbol A']
    symbol_b = row['Symbol B']
    
    if symbol_a not in connections:
        connections[symbol_a] = []
    if symbol_b not in connections:
        connections[symbol_b] = []
    
    if symbol_b not in connections[symbol_a]:
        connections[symbol_a].append(symbol_b)
    if symbol_a not in connections[symbol_b]:
        connections[symbol_b].append(symbol_a)

# Save the dictionary to a JSON file
with open('chaperone-coexpression.json', 'w') as f:
    json.dump(connections, f, indent=4)

# Print the dictionary to verify
print(json.dumps(connections, indent=4))
```
