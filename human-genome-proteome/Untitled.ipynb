```python
import pandas as pd

# Define the path to the CSV file
file_path = '/Users/b/1000-networks/human-disease/genemap2.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the required columns
extracted_df = df[['Gene/Locus And Other Related Symbols', 'Phenotypes']]

# Remove rows with empty values in either column
filtered_df = extracted_df.dropna(subset=['Gene/Locus And Other Related Symbols', 'Phenotypes'])

# Create the gene: phenotypes dictionary
gene_phenotypes = {}

# Populate the dictionary
for index, row in filtered_df.iterrows():
    genes = row['Gene/Locus And Other Related Symbols'].split(', ')
    phenotype = row['Phenotypes']
    for gene in genes:
        if gene in gene_phenotypes:
            gene_phenotypes[gene].append(phenotype)
        else:
            gene_phenotypes[gene] = [phenotype]

# Print the size of the dictionary
print(len(gene_phenotypes))

# Optionally, display the first few entries of the dictionary for verification
for gene, phenotypes in list(gene_phenotypes.items())[:5]:
    print(f"{gene}: {phenotypes}")
```

```python
import pandas as pd

# Define the path to the CSV file
file_path = '/Users/b/1000-networks/human-disease/genemap2.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the required columns
extracted_df = df[['Gene/Locus And Other Related Symbols', 'Phenotypes']]

# Remove rows with empty values in either column
filtered_df = extracted_df.dropna(subset=['Gene/Locus And Other Related Symbols', 'Phenotypes'])

# Create the gene: phenotypes dictionary without redundancy
gene_phenotypes = {}
unique_genes = set()

# Populate the dictionary
for index, row in filtered_df.iterrows():
    genes = row['Gene/Locus And Other Related Symbols'].split(', ')
    phenotype = row['Phenotypes']
    for gene in genes:
        if gene not in list(gene_phenotypes.values()):
            gene_phenotypes[gene].append(phenotype)
        else if gene not in unique_genes:
            unique_genes.add(gene)
            if gene in gene_phenotypes:
                gene_phenotypes[gene].append(phenotype)
            else:
                gene_phenotypes[gene] = [phenotype]
            break

# Print the size of the dictionary
print(len(gene_phenotypes))

# Optionally, display the first few entries of the dictionary for verification
for gene, phenotypes in list(gene_phenotypes.items())[:5]:
    print(f"{gene}: {phenotypes}")
```
